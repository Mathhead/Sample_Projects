```python
metadata_df = pd.read_csv( path_to_movie_data + "movies_metadata.csv" )

metadata_df.head()
```

Next, we combine both dataframes to a custom dataframe.<br>
For this, we match both dataframes by the __id__, and then we insert into the new dataframe the keywords from __keywords_df__ and the __original_title__, the __overview__ and the __vote_average__.

```python
def get_metadata_by_id( movie_id ):
    
    """ 
        This function takes the id as input parameter and then selects that entry in metadata_df
        which has the same movie_id. Then the data from the columns "original_title", "overview" and
        "vote_average" are extracted, put into a list, and this list is returned.
        If the movie_id cannot be found, -1 is returned.
    """
    
    df = metadata_df.loc[ metadata_df[ "id" ] == movie_id ]
    
    if df.empty:
        
        return -1
    
    
    
    overview_text = df[ "overview" ].values[ 0 ]
    
    
    
    
    if overview_text != overview_text:
        
        return -1
    
    
    
    overview_text = overview_text.lower()
    
    overview_text = overview_text.replace( ".", "" )
    
    overview_text = overview_text.replace( ",", "" )
    
    overview_text = overview_text.replace( ";", "" )
    
    overview_text = overview_text.replace( "?", "" )
    
    overview_text = overview_text.replace( "!", "" )
    
    overview_text = overview_text.replace( "-", "" )
    
    
    
    
    
    bag_of_words = overview_text.split( " " )
    
    new_bag_of_words = []
    
    for item in bag_of_words:
        
        if len( item ) > 0:
            
            if item not in new_bag_of_words:
                
                new_bag_of_words.append( item )
                
                
                
    
    data = [ df[ "original_title" ].values[ 0 ], new_bag_of_words, df[ "vote_average" ].values[ 0 ] ]
    
    return data
```

```python
def extract_keywords_by_index( index ):
    
    """
        This function first checks, whether the index is valid. If not -1 is returned.
        If the index is valid, the raw keywords string is extracted.
        We then manipulate this string until a list of keywords is achieved. If this list is
        empty, we return -1, else, we return the list.
    """
    
    if index < 0:
        
        return -1
    
    if index >= len( keywords_df ):
        
        return -1
    
    
    
    raw_string = keywords_df.iloc[ index ][ "keywords" ]
    
    raw_string = raw_string[ 1 : len( raw_string ) - 1 ]
    
    list_of_substrings = raw_string.replace( "{", "}" ).split( "}" )
    
    list_of_substrings = list_of_substrings[ 1 : len( list_of_substrings ) - 1 ]
    
    
    
    
    new_list_of_substrings = []
    
    for item in list_of_substrings:
        
        if "name" in item:
            
            new_list_of_substrings.append( item )
            
            
            
            
    another_list_of_substrings = []
    
    for item in new_list_of_substrings:
        
        another_list_of_substrings.append( item.split( "name" )[ 1 ] )
        
        
        
        
    yet_another_list_of_substrings = []
    
    for item in another_list_of_substrings:
        
        yet_another_list_of_substrings.append( item[ 4 : len( item ) - 1 ] )
        
        
        
        
    final_list_of_substrings = []
    
    for item in yet_another_list_of_substrings:
        
        if len( item ) > 0:
            
            final_list_of_substrings.append( item )
            
            
            
            
    if len( final_list_of_substrings ) == 0:
        
        return -1
         
    return final_list_of_substrings
```

```python
data = []




for i in range( len( keywords_df ) ):
    
    movie_id = str( keywords_df.loc[ i ][ "id" ] )
    
    keywords = extract_keywords_by_index( i )
    
    metadata = get_metadata_by_id( movie_id )
    
    
    
    if keywords == -1:
        
        continue
        
    insert = []
    
    if metadata == -1:
        
        continue
        
    for item in metadata:
        
        insert.append( item )
        

    
    insert.append( keywords )
    
    data.append( insert )
    
    
    
    
combined_data = pd.DataFrame( data, columns = [ "Title", "Overview", "Rating", "Keywords" ] )

combined_data.head()
```

Next, we want to compute the tf-idf-score for the words in the Overview and the Keywords.<br>
The tf-idf-score is computed as the product of the tf-score and and the idf-score.<br>
The tf-score is computed as follows: 
$$tf\left( w, D\right)=\frac{\# w \text{ in D}}{\text{max}_{w^{*} \text{ in D}}\#w^{*}}$$ 
for a given word w.<br>
The idf-score is computed as follows:
$$idf\left( w \right) = \log \left( \frac{N}{\sum_{D}1_{w\text{ in D}}}\right)$$
Here, $N$ is the number of documents, and $1_{w\text{ in D}}$ is the function that returns $1$ if $w$ is in $D$ and $0$ otherwise.<br>
In our case, each word appears only once, therefore, we only consider the idf-score.<br>
Next, we create for the Overview and the Keywords dictionaries in order to count how often each word appears.

```python
overview_word_counter = {}

keywords_word_counter = {}



for i in range( len( combined_data ) ):
    
    overview_words = combined_data.loc[ i ][ "Overview" ]
    
    for word in overview_words:
        
        if word not in overview_word_counter:
            
            overview_word_counter[ word ] = 1
            
        else:
            
            overview_word_counter[ word ] = overview_word_counter[ word ] + 1
            
            
            
    keywords_words = combined_data.loc[ i ][ "Keywords" ]
    
    for word in keywords_words:
        
        if word not in keywords_word_counter:
            
            keywords_word_counter[ word ] = 1
            
        else:
            
            keywords_word_counter[ word ] = keywords_word_counter[ word ] + 1
```

Now, we create the actual recommender.<br>
We give this function the name of a movie, and compare for all the other movies, which words from the overview and the keywords they have in common. Then, we sum over the idf-scores of these words, and finally, we multiply this sum with the rating of the other movie.

```python
import numpy as np

print( np.__version__ )
```

```python
def recommender( movie ):
    """
        First, check, whether movie appears in the data set.
        Then extract the overview and the keywords of that movie.
        Then we iterate overall the other movies. We extract their overviews and keywords, 
        and check, which of these words are in common with the words that describe movie.
        Then we sum the idf-scores of these words and multiply the result by the rating.
    """
    
    found = False
    
    j = -1
    
    N = len( combined_data )
    
    
    
    
    for i in range( N ):
        
        if movie == combined_data.loc[ i ][ "Title" ]:
            
            found = True
            
            j = i
            
            break
            
    if not found:
        
        return -1
    
    
    
    
    original_overview = combined_data.loc[ i ][ "Overview" ]
    
    original_keywords = combined_data.loc[ i ][ "Keywords" ]
    
    data = []
    
    
    
    
    for i in range( N ):
        
        if i != j:
            
            new_overview = combined_data.loc[ i ][ "Overview" ]
            
            new_keywords = combined_data.loc[ i ][ "Keywords" ]
            
            
            
            
            common_overview = []
            
            common_keywords = []
            
            
            
            
            
            for word in original_overview:
                
                if word in new_overview:
                    
                    common_overview.append( word )
                    
                    
                    
                    
                    
            for word in original_keywords:
                
                if word in new_keywords:
                    
                    common_keywords.append( word )
                    
                    
                    
                    
            
            weight = 0.0
            
            fN = np.float( N )
            
            
            
            
            
            for word in common_overview:
                
                weight = weight + np.log( fN / np.float( overview_word_counter[ word ] ) )
                
            for word in common_keywords:
                
                weight = weight + np.log( fN / np.float( keywords_word_counter[ word ] ) )  
                
                
                
                
                
            weight = weight * combined_data.loc[ i ][ "Rating" ]
            
            
            
            
            if weight > 0:
            
                insert = [ combined_data.loc[ i ][ "Title" ], weight ]
            
                data.append( insert )
            
            
            
            
            
            
    result_df = pd.DataFrame( data, columns = [ "Title", "Weight" ] )
    
    result_df = result_df.sort_values( by = "Weight", axis = 0, ascending = False )
    
    result_df = result_df.reset_index( drop = True )
    
    
    

    
    suggestions = []
    
    for i in range( min( 10, len( result_df ) ) ):
        
        suggestions.append( result_df.loc[ i ][ "Title" ] )
        
    for item in suggestions:
        
        print( item )
```
